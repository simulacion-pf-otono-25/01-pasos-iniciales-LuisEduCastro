```python
import numpy as np
import matplotlib.pyplot as plt

# Limites de integracion
a = 0
b = 2

def funcion_a_integrar(x):
    return x**2

N = 100000 # Numero de puntos

x = np.random.uniform(a, b, N) # Puntos aleatorios en [a,b]
y = np.random.uniform(funcion_a_integrar(a), funcion_a_integrar(b), N)

print(x)
print(y)

n_bajo_curva = y < funcion_a_integrar(x)
respuesta = (b-a) * funcion_a_integrar(b) * n_bajo_curva / N
print(respuesta)

plt.plot(x, y, 'o')
plt.plot(np.linspace(a, b, 10), funcion_a_integrar(np.linspace(a, b, 10)))
plt.show()  
```
